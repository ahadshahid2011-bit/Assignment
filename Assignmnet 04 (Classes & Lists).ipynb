```python
import math

class Circle:
    # constructor
    def __init__(self, r):
        self.radius = r

    # method to calculate area
    def area(self):
        return 3.14 * self.radius * self.radius

    # method to calculate perimeter
    def perimeter(self):
        return 2 * 3.14 * self.radius


# testing the class
r = float(input("Enter radius of circle: "))
c1 = Circle(r)
print("Area of circle:", c1.area())
print("Perimeter of circle:", c1.perimeter())
```

```python
class Person:
    def __init__(self, name, year):
        self.name = name
        self.year = year
    
    def age(self):
        return 2026 - self.year

p = Person("Ahad", 2011)
print("Name:", p.name)
print("Age:", p.age())
```

```python
class Calculator:
    def add(self,a,b): return a+b
    def sub(self,a,b): return a-b
    def mul(self,a,b): return a*b
    def div(self,a,b): return a/b

c = Calculator()
print(c.add(10,2))
print(c.sub(10,9))
print(c.mul(10,1))
```

```python

numbers = [24, 7, 89, 42, 15]

# Initialize largest and smallest with the first item in the list
largest = numbers[0]
smallest = numbers[0]

# Loop through the list to compare each number
for num in numbers:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

print(f"List: {numbers}")
print(f"Smallest number: {smallest}")
print(f"Largest number: {largest}")
```

```python
numbers = [42, 12, 89, 7, 25]

# Method 1: Using sorted() (creates a new list, leaves original unchanged)
ascending_list = sorted(numbers)
descending_list = sorted(numbers, reverse=True)

print("Original List:  ", numbers)
print("Ascending Order: ", ascending_list)
print("Descending Order:", descending_list)
```

```python
# Nested list of students
students = [
    ["Ali", 20],
    ["Ahmed", 18],
    ["Sara", 19]
]

# Print all student names
print("Student Names:")
for student in students:
    print(student[0])

print()  # Empty line for clean spacing

# Print all student ages
print("Student Ages:")
for student in students:
    print(student[1])
```

```python
# Create a list of 10 numbers
numbers = [15, 8, 42, 23, 16, 4, 30, 91, 55, 67]

# Perform calculations
total_numbers = len(numbers)
total_sum = sum(numbers)
average = total_sum / total_numbers
max_value = max(numbers)
min_value = min(numbers)

# Print results
print(f"Total numbers: {total_numbers}")
print(f"Sum:           {total_sum}")
print(f"Average:       {average}")
print(f"Maximum value: {max_value}")
print(f"Minimum value: {min_value}")
```

```python
# Initial list
fruits = ["Apple", "Banana", "Cherry"]
print("Initial list: ", fruits)

# 1. Append one item (adds to the end)
fruits.append("Orange")
print("After append: ", fruits)

# 2. Insert one item (adds at a specific index, e.g., index 1)
fruits.insert(1, "Mango")
print("After insert: ", fruits)

# 3. Remove one item (removes the first matching value)
fruits.remove("Banana")
print("After remove: ", fruits)

# 4. Pop one item (removes and returns item at given index, default is last)
popped_item = fruits.pop()
print("After pop:    ", fruits, f" (Popped: '{popped_item}')")
```
