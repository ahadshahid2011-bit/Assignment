```python
a, b = 10, 5

# Arithmetic
print(a+b, a-b, a*b, a/b, a//b, a%b, a**b)

# Comparison
print(a==b, a!=b, a>b, a<b)

# Logical
print(a>5 and b<10, a>5 or b>10, not a==b)

# Assignment
a += 2; b -= 1; print(a, b)

# Bitwise
print(a & b, a | b, a ^ b, ~a, a << 1, a >> 1)

# Membership + Identity
lst = [1,2,3]
print(2 in lst, 5 not in lst)
x = y = 10
print(x is y, x is not a)

# Ternary
print("Big" if a>b else "Small")
```
