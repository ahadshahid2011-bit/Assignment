```python

salary = float(input("Enter your salary: "))
year = int(input("Enter year of service: "))

if year > 5:
    bonus = salary * 5 / 100
    net_salary = salary + bonus
    print("Your bonus is:", bonus)
    print("Net salary after bonus:", net_salary)
else:
    print("No bonus. Net salary:", salary)
```

```python
marks = int(input("Enter marks: "))

if marks < 25:
    print("Grade: F")
elif marks <= 45:
    print("Grade: E")
elif marks <= 50:
    print("Grade: D")
elif marks <= 60:
    print("Grade: C")
elif marks <= 80:
    print("Grade: B")
else:
    print("Grade: A")
```
